```python
import json
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
```

```python

vc = pd.read_excel("Crunshbase.xlsx")
vc['first_name'] = vc['first_name'].str.lower()
vc['last_name'] = vc['last_name'].str.lower()
```

```python

wiki_urls = {"first_name":[], "last_name":[], "forbes_url":[], "d1":[],"d2":[], "d3":[], "d4":[], "d5":[], "d6":[]}
def wiki_links():
    for x,y in zip(vc['first_name'],vc['last_name']):
        wiki_urls["forbes_url"].append("https://www.forbes.com/profile/"+ x + "-" + y + '/')
        wiki_urls["first_name"].append(x)
        wiki_urls["last_name"].append(y)
        
wiki_links()
```

```python
r = []
v = []

for i in wiki_urls["forbes_url"]:
    r = requests.get(i)
    if str(r) == "<Response [200]>":
        v = BeautifulSoup(r.content, 'html.parser')
        v = v.find_all("div", class_="value")
        wiki_urls["d1"].append(v[0].string)
        wiki_urls["d2"].append(v[1].string)
        wiki_urls["d3"].append(v[2].string)
        wiki_urls["d4"].append(v[3].string)
        try: 
            wiki_urls["d5"].append(v[4].string)
        except IndexError:  
            wiki_urls["d5"].append(0)
        try: 
            wiki_urls["d6"].append(v[5].string)
        except IndexError:  
            wiki_urls["d6"].append(0)
            
    else:   
        
        wiki_urls["d1"].append(0)
        wiki_urls["d2"].append(0)
        wiki_urls["d3"].append(0)
        wiki_urls["d4"].append(0)
        wiki_urls["d5"].append(0)
        wiki_urls["d6"].append(0)

```

```python
Forbes = pd.DataFrame(wiki_urls)
Forbes
```

```python
Forbes.to_excel("Forbes.xlsx")
```
